# Data Load

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

train = "/content/drive/MyDrive/09_Thinkful/03_Capstone-Projects/04_Final_Capstone/satellite-images-of-hurricane-damage/train_another"
valid = "/content/drive/MyDrive/09_Thinkful/03_Capstone-Projects/04_Final_Capstone/satellite-images-of-hurricane-damage/validation_another"
categories = ['damage', 'no_damage']
```

### *training data*

### *validation data*

```python
for category in categories:
  path = os.path.join(valid, category) 
  for img in os.listdir(path):
    img_array = cv2.imread(os.path.join(path,img))
    plt.imshow(img_array)
    plt.show()
    break
  break
```

### *Import training data*

```python
training_data = []

def create_training_data():
  for category in categories:
    path = os.path.join(train, category)
    class_num = categories.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path,img))
        new_array = cv2.resize(img_array, (img_size, img_size))
        training_data.append([new_array, class_num])
      except Exception as e:
        pass

create_training_data()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
```

```python
for sample in training_data[:10]:
  print(sample[1])
```

```python
X = []
y = []
```

```python
for features, label in training_data:
  X.append(features)
  y.append(label)
```

```python
X
```

```python
y
```

```python
X = np.array(X).reshape(-1, img_size, img_size, 3)
```

```python
import pickle

pickle_out = open('X.pickle', 'wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('y.pickle', 'wb')
pickle.dump(y, pickle_out)
pickle_out.close()
```

### *Importing validation data*

```python
validation_data = []

def create_validation_data():
  for category in categories:
    path = os.path.join(valid, category)
    class_num = categories.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path,img))
        new_array = cv2.resize(img_array, (img_size, img_size))
        validation_data.append([new_array, class_num])
      except Exception as e:
        pass

create_validation_data()
```

```python
random.shuffle(validation_data)
```

```python
X_val = []
y_val = []
```

```python
for features, label in validation_data:
  X_val.append(features)
  y_val.append(label)
```

```python
X_val = np.array(X_val).reshape(-1, img_size, img_size, 3)
```

```python
X_val
```

```python
y_val
```

```python
import pickle

pickle_out = open('X_val.pickle', 'wb')
pickle.dump(X_val, pickle_out)
pickle_out.close()

pickle_out = open('y_val.pickle', 'wb')
pickle.dump(y_val, pickle_out)
pickle_out.close()
```

# Model Building

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv3D, MaxPooling2D, Conv2D
import pickle
```

```python
X = pickle.load(open('X.pickle', 'rb'))
y = pickle.load(open('y.pickle', 'rb'))

X_val = pickle.load(open('X_val.pickle', 'rb'))
y_val = pickle.load(open('y_val.pickle', 'rb'))
```

```python
X = X/255.0
X_val = X_val/255
```

```python
model= Sequential()
```

```python
X.shape[1:]
```

```python
model.add(Conv2D(64, 3, activation='relu', input_shape=X.shape))
model.add(Conv2D(64, 3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
import numpy as np

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
model.fit(X,np.array(y), batch_size=32, validation_split=0.1, epochs=20)
```
